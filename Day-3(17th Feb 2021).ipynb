```python
# prime number
# 3-->1,3 -->count-2
# 17-->1,17

def isprime(n):
    c=0
    for i in range(1,n+1):
        if n%i==0:
            c=c+1
    if c==2:
        return True
    else:
        return False
isprime(16)
            
```

```python
for i in range(1,100):
    if isprime(i):
        print(i,end=" ")
```

```python
#perfect Number
# 6
# 1+2+3-->6
# 28
# 1+2+4+7+14-->28
def isperfect(n):
    s=0
    for i in range(1,n):
        if n%i==0:
            s=s+i
    if n==s:
        return True
    else:
        return False
isperfect(30)
```

```python
for i in range(1,100001):
    if isperfect(i):
        print(i,end=" ")
```
